```python
transport_data = [
    {"station_id": "ST001", "line_name": "Central", "journey_count": 100},
    {"station_id": "ST002", "line_name": "East", "journey_count": 200},
    {"station_id": "ST003", "line_name": "West", "journey_count": 300},
    {"station_id": "ST004", "line_name": "Nord", "journey_count": 300},
]
```

```python
df = spark.createDataFrame(transport_data)
display(df)
```

```python
from pyspark.sql import functions as F

summaty_df = df.groupBy("line_name").agg(F.sum("journey_count").alias("total_journeys"))
display(summaty_df)
```
